```python
import random
while True:
  rand_num = random.randint(1,6)
  print(rand_num)
  if rand_num == 4:
    break
  print(’.’)
```

```python
import random
while True:
  rand_num = random.randint(1,6)
  print(rand_num)
  if rand_num == 4:
        break
        print('.')
```

breakによって無限ループを強制的に止める

```python
import random
while True:
    rand_num = random.randint(1,6)
    print(rand_num)
    if rand_num == 4:
        continue
        print()
```
